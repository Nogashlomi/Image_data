```python
import practicalSPARQL
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import practicalSPARQL
import numpy as np
```

```python
root_q = 'queries'
root_d = 'results'

# read login data
login = pd.read_json('config_sparql.json')
ENDPOINT = login['endpoint'][0]
USERNAME = login['username'][0]
PASSWORD = login['password'][0]

# create sparql object
sparql = practicalSPARQL.practicalWrapper(ENDPOINT)
sparql.setCredentials(USERNAME, PASSWORD)

print("--- Querying ENDPOINT: {} ---".format(ENDPOINT))
```

```python
path_to_xlsx = 'G:\\My Drive\\DataFiles\\Full_Image_List_6.5.24\\all_elements.xlsx'
```

```python
df = pd.read_excel(path_to_xlsx)
```

```python
#all adaption of 100

q = practicalSPARQL.stringify_SPARQL('adaptions_100.sparql')
df_adaptions_100 = sparql.select_as_dataframe(q) 
```

```python
#all influenced by 100

q = practicalSPARQL.stringify_SPARQL('influenced_100.sparql')
df_influenced_100 = sparql.select_as_dataframe(q) 
```

```python
# adding a part-type column to df 

#Define the value to check for part-type: 100
specific_value = "http://sphaera.mpiwg-berlin.mpg.de/id/part/c6145ca3-317c-4724-a344-3c2e106fe26b"

# Convert the columns to sets for faster lookup
adaptions_set = set(df_adaptions_100['adaption'])
influenced_set = set(df_influenced_100['influenced_part'])

# Define the function to determine the part type
def determine_part_type(part_or_adaption):
    if part_or_adaption == specific_value:
        return '100'
    elif part_or_adaption in adaptions_set:
        return 'adaption_100'
    elif part_or_adaption in influenced_set:
        return 'influenced_100'
    else:
        return 'other'

# Apply the function to the 'part_or_adaption' column
df['part type'] = df['part_or_adaption'].apply(determine_part_type)
```

```python
# Print unique values of "part type"
unique_part_types = df['part type'].unique()
print("Unique values of 'part type':", unique_part_types)
```

```python
#ploting the number of images per part per book by part type in every 20 years
# Create the 20-year interval column
bins = np.arange(1470, 1670, 20)
labels = [f'{i}-{i+19}' for i in bins[:-1]]
df.loc[:, '20_year_interval'] = pd.cut(df['year'], bins=bins, labels=labels, right=False)

# Group by the 20-year intervals, book, part, and part type, then count the number of images
grouped = df.groupby(['20_year_interval', 'book', 'part_or_adaption', 'part type'], observed=True).size().reset_index(name='count')

# Aggregate the counts per 20-year interval and part type
aggregated = grouped.groupby(['20_year_interval', 'part type']).agg({'count': 'sum'}).reset_index()

# Plotting the data
plt.figure(figsize=(15, 10))
sns.set(style="whitegrid")

# Create a line plot
sns.lineplot(data=aggregated, x='20_year_interval', y='count', hue='part type', marker='o')

# Customize the plot
plt.title('Number of Images per Book per Part in 20-Year Intervals by Part Type')
plt.xlabel('20-Year Interval')
plt.ylabel('Number of Images')
plt.legend(title='Part Type')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
