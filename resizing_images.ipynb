```python
import os
from PIL import Image as PILImage

# Path to your original image folder
original_image_folder = r'C:\\Users\\nogas\\Image_data\\images_ck_flat'

# Path to the new folder where resized images will be saved
resized_image_folder = r'C:\\Users\\nogas\\Image_data\\images_ck_flat_resized'

# Create the new folder if it doesn't exist
if not os.path.exists(resized_image_folder):
    os.makedirs(resized_image_folder)

# Resize each image and save it to the new folder
for filename in os.listdir(original_image_folder):
    if filename.lower().endswith('.jpg'):
        # Construct full image path
        original_image_path = os.path.join(original_image_folder, filename)
        
        # Open the original image
        with PILImage.open(original_image_path) as img:
            # Resize the image to a smaller size, e.g., 100x100 pixels
            img_resized = img.resize((10, 10))  # Adjust the size as needed
            
            # Save the resized image to the new folder
            resized_image_path = os.path.join(resized_image_folder, filename)
            img_resized.save(resized_image_path)

        print(f"Resized and saved: {filename}")

print("All images have been resized and saved.")
```
