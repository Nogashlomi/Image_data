```python
import practicalSPARQL
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
root_q = 'queries'
root_d = 'results'

# read login data
login = pd.read_json('config_sparql.json')
ENDPOINT = login['endpoint'][0]
USERNAME = login['username'][0]
PASSWORD = login['password'][0]

# create sparql object
sparql = practicalSPARQL.practicalWrapper(ENDPOINT)
sparql.setCredentials(USERNAME, PASSWORD)

print("--- Querying ENDPOINT: {} ---".format(ENDPOINT))
```

```python
q = practicalSPARQL.stringify_SPARQL('elements_query_050824.sparql')    # select data from the ttl file as a dataframe
df = sparql.select_as_dataframe(q)
```

```python
q = practicalSPARQL.stringify_SPARQL('books_query.sparql')    # select data from the ttl file as a dataframe
books = sparql.select_as_dataframe(q)
```

```python
colures = df[df['cks'].str.contains('CK_Colures', na=False)]
```

```python
# Path to the Excel file
excel_file_path = 'C:\\Users\\nogas\\Image_data\\corrected_celestial_circles.xlsx'

# Load the Excel file into a DataFrame
excel_data = pd.read_excel(excel_file_path)

colures_visual = pd.merge(colures, excel_data, on='cluster_name', how='left')

colures_visual
```

```python
# Path to the Excel file
excel_file_path = 'C:\\Users\\nogas\\Image_data\\celestial_circles.xlsx'

# Load the Excel file into a DataFrame
excel_data_1 = pd.read_excel(excel_file_path)

colures_visual_full = pd.merge(colures_visual, excel_data_1, on='cluster_name', how='left')

colures_visual_full
```

```python
# Assuming 'colures_visual' is your DataFrame and 'books' is your books DataFrame
# Convert 'year' to int if not already done
colures_visual['year'] = colures_visual['year'].astype(int)
books['year'] = books['year'].astype(int)

# Create 20-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', 
          '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Create a new column for the 20-year intervals
colures_visual['interval'] = pd.cut(colures_visual['year'], bins=bins, labels=labels, right=False)

# Count unique images for "3D decoration" where value is "yes"
unique_images_decoration = colures_visual[colures_visual['3D decoration'] == 'yes'] \
    .groupby('interval')['images'].nunique().reindex(labels, fill_value=0)

# Count unique images for "3D main point" where value is "yes"
unique_images_main_point = colures_visual[colures_visual['3D main point'] == 'yes'] \
    .groupby('interval')['images'].nunique().reindex(labels, fill_value=0)

# Create a new column for the 20-year intervals in books DataFrame
books['interval'] = pd.cut(books['year'], bins=bins, labels=labels, right=False)

# Count unique books per interval
unique_books_count = books.groupby('interval')['book'].nunique().reindex(labels, fill_value=0)

# Plotting
plt.figure(figsize=(12, 7))
plt.plot(labels, unique_images_decoration, marker='o', color='cornflowerblue', label='3D Decoration')
plt.plot(labels, unique_images_main_point, marker='o', color='lightcoral', label='3D Main Point')
plt.plot(labels, unique_books_count, marker='o', color='darkgreen', label='Unique Books', linestyle='--')

# Formatting the plot
plt.xlabel('Year Intervals')
plt.ylabel('Count')
plt.title('Unique Images and Books by 3D Decoration, Main Point, and Year Intervals (1470-1650)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
