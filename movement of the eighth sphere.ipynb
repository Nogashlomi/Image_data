```python
# Step 1: Filter `all_elements_with_uck` by the specific `uck` value
uck_value = 'Movement of the Eighth Sphere'  # Replace with your specific uck value
filtered_df_uck = all_elements_with_uck[all_elements_with_uck['uck'] == uck_value]

# Step 2: Extract the `label` values
labels = filtered_df_uck['label'].tolist()

# Step 3: Filter `all_elements_full_info` using the extracted `label` values to match `cluster_name`
Movement_Eighth_Sphere = all_elements_full_info[all_elements_full_info['cluster_name'].isin(labels)]
```

```python
# Define the twenty-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Define the year intervals
Movement_Eighth_Sphere['year_interval'] = pd.cut(Movement_Eighth_Sphere['year'], bins=bins, labels=labels, right=False)

# Group by year_interval and part_type, then count unique images
grouped_df = Movement_Eighth_Sphere.groupby(['year_interval', 'part_type'], observed=True).agg({'images': 'nunique'}).reset_index()

# Pivot the DataFrame to have part_type as columns and year_interval as rows
pivot_df = grouped_df.pivot(index='year_interval', columns='part_type', values='images').fillna(0)

# Plotting
ax = pivot_df.plot(kind='line', marker='o', figsize=(12, 8))
plt.title('Number of Unique Images by Part Type and Year Interval')
plt.xlabel('Year Interval')
plt.ylabel('Number of Unique Images')
plt.xticks(rotation=45)
plt.legend(title='Part Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```
