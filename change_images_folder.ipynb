```python
import os
import shutil

# Path to the main image directory
base_dir = 'C:\\Users\\nogas\\Image_data\\images_ck'  # Current folder with subfolders
destination_dir = 'C:\\Users\\nogas\\Image_data\\images_ck_flat'  # The folder where all images will be copied

# Step 1: Create the destination folder if it doesn't exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Step 2: Create a list of all subdirectories (which represent the keywords)
subdirectories = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

# Step 3: Loop over each subdirectory (keyword folder)
for subdir in subdirectories:
    # Get the full path to the subdirectory
    subdir_path = os.path.join(base_dir, subdir)
    
    # Step 4: List all the image files in this subdirectory
    image_files = [f for f in os.listdir(subdir_path) if f.endswith('.jpg')]  # Assuming .jpg format

    # Step 5: Copy each image to the destination folder, keeping original filenames
    for image_file in image_files:
        # Define the full path to the image in the source folder
        old_image_path = os.path.join(subdir_path, image_file)
        
        # Define the path where the image will be copied in the destination folder
        new_image_path = os.path.join(destination_dir, image_file)
        
        # Step 6: Check if the image already exists in the destination folder
        if os.path.exists(new_image_path):
            print(f"Duplicate found, skipping: {old_image_path}")
        else:
            # Step 7: Copy the image to the destination folder
            shutil.copy2(old_image_path, new_image_path)  # Copy the file and preserve metadata
            print(f"Copied: {old_image_path} -> {new_image_path}")

# Optional: You can now inspect the flattened folder (destination_dir) for all images
```
