```python
import pandas as pd 
import glob
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
all_elements_full_info = pd.read_excel(path_to_xlsx)
```

```python
path_to_csv = 'G:\\My Drive\\DataFiles\\booksSacrobosco.csv'
```

```python
all_books = pd.read_csv(path_to_csv)
```

```python
heavens_spherical = all_elements_full_info[all_elements_full_info['keywords'].str.contains('CK_Sphericity of the Heavens')]
```

```python
# Define the 20-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Create a new column 'year_interval' to hold the 20-year intervals in water_earth
heavens_spherical['year_interval'] = pd.cut(heavens_spherical['year'], bins=bins, labels=labels, right=True)

# Group by year_interval and part_type, and count unique images
grouped = heavens_spherical.groupby(['year_interval', 'part_type'], observed=True)['images'].nunique().reset_index()

# Pivot the table for easier plotting
pivot_df = grouped.pivot(index='year_interval', columns='part_type', values='images').fillna(0)

# Create a new column 'year_interval' in all_books
all_books['year_interval'] = pd.cut(all_books['year_published_from'], bins=bins, labels=labels, right=True)

# Group by year_interval and count unique subjects
grouped_subjects = all_books.groupby('year_interval', observed=True)['subject'].nunique().reset_index()

# Plot the data
plt.figure(figsize=(12, 6))

# Plot unique images for each part_type
for part_type in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[part_type], marker='o', label=f'Part Type {part_type}')

# Plot unique subjects
plt.plot(grouped_subjects['year_interval'], grouped_subjects['subject'], marker='o', linestyle='--', color='black', label='Toal Printed Books in the Corpus')

plt.title('Sphericity of the Heavens')
plt.xlabel('Year Interval')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Function to extract unique CK values
def extract_unique_ck_values(column):
    # Split values and filter those starting with 'CK'
    ck_values = column.str.split(';').explode()
    ck_values = ck_values[ck_values.str.startswith('CK')]
    # Get unique CK values
    unique_ck_values = ck_values.unique()
    return unique_ck_values

unique_ck_values = extract_unique_ck_values(heavens_spherical['keywords'])

print(unique_ck_values)
```

```python
# Define the specific keywords of interest
specific_keywords = [
    'CK_Isoperimetry', 'CK_Apparent Size of Stars', 'CK_Refraction'
]

# Create a function to check if any of the specific keywords is in the 'keywords' column
def contains_specific_keyword(keywords_list, keyword):
    return keyword in keywords_list

# Add a column for each specific keyword to check if it is in the 'keywords' list
for keyword in specific_keywords:
    heavens_spherical[f'has_{keyword}'] = heavens_spherical['keywords'].apply(
        lambda x: contains_specific_keyword(x.split(';'), keyword)
    )

# Add a column for images that do not have any of the specified keywords
heavens_spherical['has_none_of_keywords'] = ~heavens_spherical['keywords'].apply(
    lambda x: any(keyword in x.split(';') for keyword in specific_keywords)
)

# Create bins and labels for 20-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Assign each row to a 20-year interval
heavens_spherical['interval'] = pd.cut(heavens_spherical['year'], bins=bins, labels=labels, right=False)

# Plotting
plt.figure(figsize=(14, 10))

# Total unique images
total_unique_images = heavens_spherical.groupby('interval')['images'].nunique()
plt.plot(total_unique_images.index, total_unique_images.values, label='Total Unique Images', color='blue')

# Plot each specific keyword line
for keyword in specific_keywords:
    keyword_column = f'has_{keyword}'
    unique_images_with_keyword = heavens_spherical[heavens_spherical[keyword_column]]\
        .groupby('interval')['images']\
        .nunique()
    plt.plot(unique_images_with_keyword.index, unique_images_with_keyword.values, label=keyword)

# Plot the line for images without any of the specific keywords
unique_images_without_keywords = heavens_spherical[heavens_spherical['has_none_of_keywords']]\
    .groupby('interval')['images']\
    .nunique()
plt.plot(unique_images_without_keywords.index, unique_images_without_keywords.values, label='Without Keywords', linestyle='--', color='black')

plt.xlabel('Year Interval')
plt.ylabel('Number of Unique Images')
plt.title('Number of Unique Images by Keyword and 20-Year Interval')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Define 20-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']
heavens_spherical['period'] = pd.cut(heavens_spherical['year'], bins=bins, labels=labels, right=False)

# Aggregate data for unique images and unique books
images_per_place = heavens_spherical.groupby(['place', 'period']).agg(
    unique_images=('images', pd.Series.nunique), 
    unique_books=('book', pd.Series.nunique)
).reset_index()

# Handle division by zero
images_per_place['normalized_unique_images'] = images_per_place.apply(
    lambda row: row['unique_images'] / row['unique_books'] if row['unique_books'] != 0 else 0, axis=1
)

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))
for place in images_per_place['place'].unique():
    place_data = images_per_place[images_per_place['place'] == place]
    ax.plot(place_data['period'], place_data['normalized_unique_images'], marker='o', label=place)

ax.set_title('Normalized Unique Images per Book by Place and Period')
ax.set_xlabel('Period')
ax.set_ylabel('Normalized Unique Images')
ax.legend(title='Place', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Prepare data for clustering
features = images_per_place[['normalized_unique_images', 'unique_books']].fillna(0)
kmeans = KMeans(n_clusters=5, random_state=0).fit(features)
images_per_place['cluster'] = kmeans.labels_

# Plotting clusters
fig, ax = plt.subplots(figsize=(12, 8))
for cluster in images_per_place['cluster'].unique():
    cluster_data = images_per_place[images_per_place['cluster'] == cluster]
    ax.plot(cluster_data['period'], cluster_data['normalized_unique_images'], marker='o', label=f'Cluster {cluster}')

ax.set_title('Clusters of Cities Based on Image and Book Data')
ax.set_xlabel('Period')
ax.set_ylabel('Normalized Unique Images')
ax.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
