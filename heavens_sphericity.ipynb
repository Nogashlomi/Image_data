```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming merged_df is your DataFrame
# Ensure the 'year' column is of type int
merged_df['year'] = merged_df['year'].astype(int)

# Create 20-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Create a new column for the 20-year intervals
merged_df['interval'] = pd.cut(merged_df['year'], bins=bins, labels=labels, right=False)

# Filter the DataFrame for "human figure" == "yes"
human_figures_df = merged_df[merged_df['human figure'] == 'yes']

# Group by interval and count unique images with "human figure" == "yes"
unique_images_human_figures = human_figures_df.groupby('interval')['images'].nunique()

# Group by interval and count total unique images
total_unique_images = merged_df.groupby('interval')['images'].nunique()

# Function to get the top three places with the most unique images for each interval
def get_top_places_per_interval(interval):
    interval_df = merged_df[merged_df['interval'] == interval]
    top_places = interval_df.groupby('place')['images'].nunique().nlargest(3)
    return top_places

# Plot the results
plt.figure(figsize=(12, 6))

plt.plot(unique_images_human_figures.index, unique_images_human_figures.values, marker='o', linestyle='-', color='blue', label='Human Figure')
plt.plot(total_unique_images.index, total_unique_images.values, marker='o', linestyle='-', color='green', label='Total Unique Images')

# Add annotations for the top three places above each dot
for interval in labels:
    if interval in unique_images_human_figures.index:
        top_places = get_top_places_per_interval(interval)
        annotation_text = '\n'.join([f'{place}: {count}' for place, count in top_places.items()])
        
        human_figures_value = unique_images_human_figures.loc[interval]
        total_images_value = total_unique_images.loc[interval]
        
        # Annotation for human figures line
        plt.annotate(annotation_text, (interval, human_figures_value), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color='blue')
        
        # Annotation for total images line
        plt.annotate(annotation_text, (interval, total_images_value), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color='green')

plt.xlabel('20-Year Interval')
plt.ylabel('Number of Unique Images')
plt.title('Unique Images by 20-Year Intervals (1470-1650)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the 20-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Convert year columns to numeric
merged_df['year'] = pd.to_numeric(merged_df['year'], errors='coerce')
books['year'] = pd.to_numeric(books['year'], errors='coerce')

# Convert 'cks' column to string
merged_df['cks'] = merged_df['cks'].astype(str)

# Add an 'interval' column to both DataFrames
merged_df['interval'] = pd.cut(merged_df['year'], bins=bins, labels=labels, right=False)
books['interval'] = pd.cut(books['year'], bins=bins, labels=labels, right=False)

# Filter merged_df for each keyword and count unique images
def count_unique_images_for_cks(cks_value):
    return merged_df[merged_df['cks'].str.contains(cks_value, na=False)].groupby('interval', observed=True)['images'].nunique()

cks_values_list = {
    'CK_Sphericity of the Heavens': 'CK_Sphericity of the Heavens',
    'CK_Refraction': 'CK_Refraction',
    'CK_Apparent Size of Stars': 'CK_Apparent Size of Stars',
    'CK_Isoperimetry': 'CK_Isoperimetry'
}

# Prepare data for plotting
plot_data = pd.DataFrame(index=labels)

for label, cks_value in cks_values_list.items():
    plot_data[label] = count_unique_images_for_cks(cks_value)

# Calculate number of unique books in each interval
book_counts = books.groupby('interval', observed=True)['book'].nunique()
plot_data['Unique Books'] = book_counts

# Calculate number of unique images that do not have the specified keywords
keywords_to_exclude = ['CK_Refraction', 'CK_Apparent Size of Stars', 'CK_Isoperimetry']
def count_unique_images_excluding_keywords(keywords):
    # Flag images that contain any of the specified keywords
    mask = merged_df['cks'].str.contains('|'.join(keywords), na=False)
    # Filter and count unique images that do not contain the specified keywords
    return merged_df[~mask].groupby('interval', observed=True)['images'].nunique()

plot_data['Images Excluding Keywords'] = count_unique_images_excluding_keywords(keywords_to_exclude)

# Plotting
plt.figure(figsize=(12, 8))

# Plot each line with default color and style
for column in plot_data.columns:
    if column == 'Unique Books':
        plt.plot(plot_data.index, plot_data[column], label=column, color='black', linestyle='--')
    else:
        plt.plot(plot_data.index, plot_data[column], label=column)

plt.xlabel('20-Year Intervals')
plt.ylabel('Count')
plt.title('Sphericity of the Heavens Images')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the twenty-year intervals
bins = [1470, 1490, 1510, 1530, 1550, 1570, 1590, 1610, 1630, 1650]
labels = ['1470-1489', '1490-1509', '1510-1529', '1530-1549', '1550-1569', '1570-1589', '1590-1609', '1610-1629', '1630-1649']

# Extract the year from 'year' column and bin it into intervals
vk_data['interval'] = pd.cut(vk_data['year'], bins=bins, labels=labels, right=False)

# Define the keywords to filter
keyword_1 = 'CK_Sphericity of the Heavens'
keyword_2 = 'VK_Ad absurdum'

# Create a boolean mask for the conditions
has_keyword_1 = vk_data['keywords'].str.contains(keyword_1)
has_keyword_2 = vk_data['keywords'].str.contains(keyword_2)
has_both_keywords = has_keyword_1 & has_keyword_2

# Calculate the number of unique images for each condition and interval
unique_images_keyword_1 = vk_data[has_keyword_1].groupby('interval')['images'].nunique()
unique_images_both_keywords = vk_data[has_both_keywords].groupby('interval')['images'].nunique()

# Initialize the plot
plt.figure(figsize=(12, 6))

# Plot unique images for each condition
plt.plot(unique_images_keyword_1.index, unique_images_keyword_1.values, label=f'Unique Images with {keyword_1}', marker='o')
plt.plot(unique_images_both_keywords.index, unique_images_both_keywords.values, label=f'Unique Images with both {keyword_1} and {keyword_2}', marker='o')

# Add annotations
def add_annotations(df, unique_images_series):
    for interval in unique_images_series.index:
        subset = df[df['interval'] == interval]
        # Calculate the top three places
        top_places = subset['place'].value_counts().head(3)
        annotations = [f"{place} ({count})" for place, count in top_places.items()]
        annotation_text = "\n".join(annotations)
        
        # Plot annotations
        y_value = unique_images_series.loc[interval]
        plt.text(interval, y_value, annotation_text, fontsize=9, ha='center', va='bottom', bbox=dict(facecolor='white', alpha=0.5))

# Add annotations for both keyword conditions
add_annotations(vk_data[has_keyword_1], unique_images_keyword_1)
add_annotations(vk_data[has_both_keywords], unique_images_both_keywords)

# Set plot labels and title
plt.xlabel('Time Interval')
plt.ylabel('Number of Unique Images')
plt.title('Number of Unique Images in Twenty-Year Intervals (1470-1650)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
